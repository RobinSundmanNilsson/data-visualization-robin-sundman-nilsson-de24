```python
import pandas as pd

df = pd.read_csv("data/1900-census-immigrant-pop-metro.csv")

df.head()
```

```python
df.info()
```

```python
df_nordic = df[["Sweden", "Norway", "Denmark", "Finland", "Total foreign born", "State"]]
df_nordic.head()
```

```python
df_nordic = df_nordic.fillna("0")
```

```python
columns = ["Sweden", "Norway", "Denmark", "Finland", "Total foreign born"]

for col in columns:
    df_nordic[col] = df_nordic[col].str.replace(",", "").astype(int)
```

```python
df_nordic.info()
```

```python
import duckdb

df_nordic_summary = duckdb.query(
    """--sql
    SELECT
        State,
        SUM(Sweden) AS Sweden,
        SUM(Norway) AS Norway,
        SUM(Denmark) AS Denmark,
        SUM(Finland) AS Finland,
    FROM
        df_nordic
    WHERE
        State IN ('Illinois', 'Minnesota', 'New York')
    GROUP BY
        State
    """
).df()
```

```python
df_nordic_summary = df_nordic_summary.set_index("State")
df_nordic_summary = df_nordic_summary.loc[["Illinois", "Minnesota", "New York"]]
df_nordic_summary
```
